# 유사한 사용자가 산 상품 추출

문제점1. unknown 제거
문제점2. 중복 제거

```python
import pickle
with open('re_clu.txt','rb') as f:
    trans_sort=pickle.load(f)
```

```python
trans_sort
```

한 사용자가 산 상품 clnt_id:product_list 형태로 변환

```python
#사용자별 산 상품 갯수 출력하여 pd_c 리스트 반환
trans_id=trans_sort['trans_id']
trans_id=list(trans_id)
count={}
for i in trans_id:
    try: count[i] += 1
    except: count[i]=1
print(count)
```

```python
values=count.values()
values=list(values)
```

```python
li=[0]
for a in range(len(values)):
    b=li[-1]+values[a]
    li.append(b)
```

```python
pd_c=list(trans_sort['pd_c'])
```

```python
# 사용자별 상품을 리스트로 묶기
pd=[]
for a in range(len(li)):
    if a+1<len(li):
        b=pd_c[li[a]:li[a+1]]
        pd.append(b)
```

```python
pd
```

```python
#문제점2. 중복제거
pd_li=[]
for a in pd:
    a=set(a)
    a=list(a)
    pd_li.append(a)
```

최종 clnt_id : pd_c

```python
key=list(count.keys())
len(key)
```

```python
trans_pd={}
for a in range(len(pd_li)):
    trans_id=key[a]
    pd_c=pd_li[a]
    trans_pd[trans_id]=pd_c
```

```python
trans_pd
```

```python
import pandas as pd
t=pd.DataFrame(list(trans_pd.items()), columns=['trans_id','pd_c'])
t
```

```python
with open('final_pd_c_list.txt','wb') as f:
    pickle.dump(t,f)
```
